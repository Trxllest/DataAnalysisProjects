```python
#Import Libraries 
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Config of plot 


#Readata

df = pd.read_csv("/Users/alpha/Downloads/movies.csv")
df = df.dropna() 
#df.dropna(how='any', axis=0, inplace=True)

```

```python
df.shape
```

```python
df.head()
```

```python
# Check for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print("{} - {}%".format(col,pct_missing))
```

```python
#Column Data Datatypes
df.dtypes
```

```python
#Changing DataType of Columns
df['budget'] = df['budget'].astype("int64")
df['gross'] = df['gross'].astype("int64")
df.head()
```

```python
#Creating a column of the correct release year
df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#drop any duplicates
#df['company'].drop_duplicates().sort_values(ascending=False)
# There were no duplicates
```

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Regression Plot for Budget vs Gross earnings using Seaborn
sns.regplot(x='budget',y='gross',data=df, scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
# Looking at correlation
df.corr() #pearson (default),kendall,spearman
```

```python
# High correlation Between budget and gross
```

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Variables')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Company Influence 

# Assign Numeric Value to Company
df_numerized = df
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == "object"):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
df.head()
```

```python
corr_matrix = df_numerized.corr()
sns.heatmap(corr_matrix,annot=True)
plt.title('Correlation Matrix for Numeric Variables')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
corr_mat = df_numerized.corr()
corr_pairs = corr_mat.unstack()
corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
